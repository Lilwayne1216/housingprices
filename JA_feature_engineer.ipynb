```python
# total other rooms bdrm_abv ground- total rooms abv ground
df['total_other_rooms']=df['TotRms AbvGrd']-df['Bedroom AbvGr']
```

```python
#how many have type 2 basement 
df['finished_2nd_bsmt']=df['BsmtFin SF 2']>0
```

```python
df['Enclosed Porch'].value_counts()
```

```python
#combine outdoor spaces 
#use boolean indexing to make outdoor space 
df['outdoorspace']=df[['Wood Deck SF','Open Porch SF','Enclosed Porch','3Ssn Porch','Screen Porch']].sum(axis=1) > 0
```

```python
#create regular lot boolean by using lot shape = 'reg'
df['reg_lot_shp']=df['Lot Shape']=='Reg'
```

```python
# create level vs not level 
df['leveled_land']=df['Land Contour']=='Lvl'
```

```python
#cul de sac or not 
df['culdsac']=df['Lot Config']=='CulDSac'
```

```python
#single family yes or no
df['1fam_hm']=df['Bldg Type']=='1Fam'
```

```python
df['House Style'].value_counts()
```

```python
#finished or unfinished attic
df['unf_attic']=(df['House Style']=='2.5Unf')|(df['House Style']=='1.5Unf')
```

```python
#hip roof yes/no
df['hip_roof']=df['Roof Style']=='Hip'
```

```python
#boolean for homes that have exterios made from VinylSd, BrKface, CemntBd
df['premium_ext']=(df['Exterior 1st']=='VinylSd')|(df['Exterior 1st']=='BrkFace')|(df['Exterior 1st']=='CemntBd')
```

```python
#drop the following columns
update_df=df.drop(columns=[
'TotRms AbvGrd', # removed since we are finding the variance using df['total_other_rooms']
'finished_2nd_bsmt',# removed since we made it boolean df['finished_2nd_bsmt']
'Bsmt Half Bath',# do not need insignificant
'Wood Deck SF',# created outdoor column df['outdoorspace']
'Open Porch SF',# created outdoor column df['outdoorspace']
'Enclosed Porch',# created outdoor column df['outdoorspace']
'3Ssn Porch',# created outdoor column df['outdoorspace']
'Screen Porch',# created outdoor column df['outdoorspace']
'Pool Area', # drop not enough data
'Misc Val',# drop not enough data
'Mo Sold',# drop not enough data
'Yr Sold',# drop not enough data
'MS Zoning',# drop not enough data
'Street',# drop not enough data
#'Alley',# drop not enough data... need to look into it, alley wasnt in this dataframe
'Lot Shape', # made boolean with df['reg_lot_shp']
'Land Contour', #made boolean with df['leveled_land']
'Utilities',# drop not enough data, only three rows had differenet values 
'Land Slope',# drop similar to ladn contour
'Lot Config',#made boolean with df['culdsac']
'Condition 1',#remove simialr info as neighborhood
'Condition 2',#remove simialr info as neighborhood
'Bldg Type',#made boolean with df['1fam_hm']
'House Style',#made boolean with df['unf_attic']
'Roof Style',#made boolean with df['hip_roof']
'Roof Matl',# drop not enough data
'Exterior 1st',#drop made boolean with df['premium_ext'] premium='VinylSd','BrkFace','CemntBd'
'Exterior 2nd'])# remove information is in extior1st 
```
